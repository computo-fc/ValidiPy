# Simulating a chaotic billiard model with interval arithmetic

In this notebook, we show how to solve the second of the well-known ["Hundred-digit challenge"](http://en.wikipedia.org/wiki/Hundred-dollar,_Hundred-digit_Challenge_problems) problems using the ValidiPy package.

**Statement of the problem:**

A point particle bounces off fixed discs of radius $\frac{1}{3}$, placed at the points of a square lattice with unit distance between neighbouring points. The particle starts at $(x,y) = (0.5, 0.1)$, heading due east with unit speed, i.e. with initial velocity $(1, 0)$. Calculate the distance from the origin of the particle at time $t=10$.

To solve this, we use a standard implementation of the billiard by treating it as a single copy of a unit cell, centred at the origin and with side length $1$, and periodic boundary conditions. We keep track of the cell that is reached in the corresponding "unfolded" version in the complete lattice.

The code is *identical* for the standard floating-point method or the method using `ValidiPy`, except for changing the initial conditions to be intervals.

```python
import numpy

import sys
sys.path.append('../src/python')  # directory for ValidiPy source

from interval import *
```

Define the disc object. Note that the *same* definition also works for a sphere in arbitrary dimensions:

```python
class Disc(object):
    def __init__(self, centre, radius):
        self.c = centre
        self.r = radius
        
    def collision_time(self, x, v):
        A = np.dot(v, v)  # 1 in principle, but varies in practice
        B = np.dot(v, x-self.c)
        C = np.dot(x-self.c, x-self.c) - self.r*self.r
        
        disc = B**2 - A*C
        
        if not disc >= 0:
            if not disc < 0:
                print "disc = ", disc
            return None
        
        return (-B - disc**0.5) / A  # suppose outside disc
    
    def normal(self, x):
        # normal to disc at point x *on boundary*
        return (x - self.c) / self.r
    
    def __repr__(self):
        return "Disc(%s, %s)" % (self.c, self.r)
```

Define a plane; again, the definition works for any number of dimensions:

```python
class Plane(object):
    def __init__(self, point, normal):
        self.c = point
        self.n = normal
        
    def collision_time(self, x, v):
        return np.dot(self.c - x, self.n) / np.dot(v, self.n)
    
    def normal(self, x):
        return self.n
```

Calculate collisions with the elements of the current cell, i.e. the 4 planes and the disc:

```python
def do_collision(x, v, current_cell):
            
    collision_times = [boundary.collision_time(x, v) for boundary in boundaries]
    
    min_collision_time = Interval(100)
    which_hit = -1
    
    for (i, collision_time) in enumerate(collision_times):
        #print i, collision_time
        if collision_time > 0:
            if collision_time < min_collision_time:
                which_hit, min_collision_time = i, collision_time
    
    new_cell = current_cell.copy()
    
    x_new = x + v*min_collision_time 

    
    if which_hit == 0:   # collides with disc
        n = boundaries[0].normal(x_new)

        v_new = v - 2*n*np.dot(n, v)  # reflejar solo si es disco; si es plano, pasar a traves

    
    else:  # collides with a plane, so periodise correctly:
        
        jump = jump_directions[which_hit]  

        new_cell += jump
        x_new -= jump  # this is supposing that we are in exactly a unit cell
        
        v_new = v.copy()
        
    return x_new, v_new, new_cell, min_collision_time, which_hit
    
```

Set up the initial conditions:

```python
def setup_initial_condition():
    
    global trajectory
    
    global x, v
    global current_cell
    
    trajectory = []

    x0 = Interval(-0.5)
    y0 = Interval(0.1)
    u0 = Interval(1)
    v0 = Interval(0)

    current_cell = np.array([1, 0])

    x = np.array([x0, y0])
    v = np.array([u0, v0])

    x, v

    total_time = Interval(0.)

    trajectory = []
    trajectory.append([xx.mid() for xx in x + current_cell])
```

Set up the unit cell boundaries: 1 disc and 4 planes:

```python
def setup_boundaries():
    global boundaries
    global radius
    global jump_directions
    
    boundaries = []

    radius = Interval(1.) / 3

    boundaries.append( Disc(np.array([0., 0.]), radius) )

    boundaries.append(Plane(np.array([-0.5, -0.5]), np.array([0., -1.])))
    boundaries.append(Plane(np.array([0.5, -0.5]), np.array([1., 0.])))
    boundaries.append(Plane(np.array([0.5, 0.5]), np.array([0., 1.])))
    boundaries.append(Plane(np.array([-0.5, 0.5]), np.array([-1., 0.])))

    jump_directions = [np.array([0,0])]  # dummy, never used
    jump_directions.extend( [boundary.normal(_).astype(int) for boundary in boundaries[1:]] )
```

```python
def run_billiard(precision, t_final):
    set_interval_precision(precision)

    setup_initial_condition()
    setup_boundaries()
    
    total_time = Interval(0)
    
    global x, v, current_cell

    step = 1

    while total_time < t_final:

        x_new, v_new, new_cell, collision_time, which_hit = do_collision(x, v, current_cell)

        if total_time + collision_time < t_final:
            total_time += collision_time
            current_cell = new_cell

            x, v = x_new, v_new

            if which_hit==0:  # hit disc
                trajectory.append([xx.mid() for xx in x + new_cell])

        else:
            collision_time = t_final - total_time

            x_new = x + v*collision_time
            x_new += current_cell

            trajectory.append([xx.mid() for xx in x_new])

            print "Position at time ", t_final, " = ", x_new
            print "Cell: ", current_cell
            
            distance = (np.dot(x_new, x_new))**0.5
            
            print "Distance from origin = ", distance
            print "Uncertainty on distance = ", distance.diam() 
            return distance.diam()

        print x[0].lo + current_cell[0], x[1].lo + current_cell[1]

        step += 1
```
