```python
import numpy as np
from intervalo_gmpy2 import *

gmpy2.get_context().precision = 100
```

```python
class Disc(object):
    def __init__(self, centre, radius):
        self.c = centre
        self.r = radius
        
    def collision_time(self, x, v):
        A = np.dot(v, v)  # 1 in principle, but varies in practice
        B = np.dot(v, x-self.c)
        C = np.dot(x-self.c, x-self.c) - self.r*self.r
        
        disc = B**2 - A*C
        
        if not disc >= 0:
            if not disc < 0:
                print "disc = ", disc
            return None
        
        return (-B - disc**0.5) / A  # suppose outside disc
    
    def normal(self, x):
        # normal to disc at point x *on boundary*
        return (x - self.c) / self.r
    
    def __repr__(self):
        return "Disc(%s, %s)" % (self.c, self.r)
```

# Lorentz gas

Use "usual" method: single square cell between $(-0.5, -0.5)$ and $(0.5, 0.5)$

```python
class Plane(object):
    def __init__(self, point, normal):
        self.c = point
        self.n = normal
        
    def collision_time(self, x, v):
        return np.dot(self.c - x, self.n) / np.dot(v, self.n)
    
    def normal(self, x):
        return self.n
```

```python
def do_collision(x, v, current_cell):
            
    collision_times = [boundary.collision_time(x, v) for boundary in boundaries]
    
    min_collision_time = Intervalo(100)
    which_hit = -1
    
    for (i, collision_time) in enumerate(collision_times):
        #print i, collision_time
        if collision_time > 0:
            if collision_time < min_collision_time:
                which_hit, min_collision_time = i, collision_time
    
#     collision_times = np.array(collision_times)
    
#     collision_times[np.not(collision_times > 0)] = np.nan  # make negatives nan so not seen by min
    
#     print "collision_times = ", collision_times
    
#     which_hit = np.argmin(collision_times)
    
    # print "which_hit = ", which_hit
    
    #collision_time = collision_times[which_hit]
    
    new_cell = current_cell.copy()
    
    x_new = x + v*min_collision_time 
    
    
    if which_hit == 0:
        n = boundaries[0].normal(x_new)

        v_new = v - 2*n*np.dot(n, v)  # reflejar solo si es disco; si es plano, pasar a traves

    
    else:
        # hit plane, so periodise
        jump = jump_directions[which_hit]  
        # print "jump = ", jump

        new_cell += jump
        x_new -= jump  # this is supposing that we are in exactly unit cell
        
        v_new = v.copy()
        
    return x_new, v_new, new_cell, min_collision_time
    
```
