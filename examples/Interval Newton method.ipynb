```python
def newton_step(f, m, deriv):
    """Works for the standard Newton method or with deriv being an interval *not* containing 0"""
    return m - ( f(m) / deriv )
```

```python
def newton_operator(f, x):
    """Interval Newton operator acting on an *interval* x. 
    Returns a list, since it may need to return 2 intervals"""
    
    if x.diam() < tol:
        return [x]
    
    m = x.mid()
    
    image, deriv_image = differentiate(f, x)
        
    if not 0 in deriv_image:
        x_new = newton_step(f, m, deriv_image)
        
        x_new = x_new.intersection(x)
        
        if x_new is None:
            return []  # no root
        
        if x_new == x:  # didn't reduce, so bisect:
            return [Intervalo(x.lo, m), Intervalo(m, x.hi)]
            
        return [ x_new ]
        
    else: # derivative interval contains 0
        
        x_new = []

        # divide into two parts excluding 0:
        deriv_image1 = Interval(deriv_image.lo, -0.0)
        deriv_image2 = Interval(0.0, deriv_image.hi)
        
        y1 = newton_step(f, m, deriv_image1)
        y1 = y1.intersection(x)
        
        y2 = newton_step(f, m, deriv_image2)
        y2 = y2.intersection(x)
        
        if y1 is not None:
            
            if y1==x:
                return [Interval(x.lo, m), Interval(m, x.hi)]
            
            else:
                x_new.append(y1)
                        
        if y2 is not None:
            if y2==x:
                return [Interval(x.lo, m), Interval(m, x.hi)]
            
            else:
                x_new.append(y2)
                
            if y1 is not None:
                if y2 < y1:
                    return [y2, y1]
            
    return x_new


def newton(f, x):
    """Do one Newton step on a *list* of intervals. 
    Returns the new list"""
    
    if not isinstance(x, list):
        x = [x]
        
    new_x = []
        
    for xx in x:
        new_x.extend(newton_operator(f, xx))
        
    return new_x


def interval_newton_method(f, I, tolerance):
    for i in range(1000):
        I = newton(f, I)
        
        diameters = [xx.diam() for xx in I]
        if max(diameters) < tol:
            break
            
    return I
```

## Check: exact calculation of periodic points for the logistic map with $r=4$

The logistic map with $r=4$ is a special case, since there is a known, exact mapping (homeomorphism) from the *tent map* to $f_4$. The tent map is sufficiently simple that all its periodic points may be calculated *analytically*, and these may then be mapped to find the analytical periodic points of the logistic map $f_4$.

The tent map is a piecewise linear map from $[0,1]$ onto itself, given by
$$T(x) := 
\begin{cases} 
2x, & \text{if } x < \frac{1}{2}; \\
2 - 2x, & \text{if } x > \frac{1}{2}.
\end{cases}
$$

```python
def tent(x):
    if x < 0.5:
        return 2*x
    return 2-2*x
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(0, 1, 1000)
y = [tent(x) for x in xx]

plt.figure(figsize=(5,5))
plt.plot(xx, y)
plt.axis("equal")
plt.xlim(0, 1)
plt.ylim(0,1)
plt.plot(xx, xx, '--')
```

```python
xx = np.linspace(0, 1, 1000)
y = [iterate(tent, 2, x) for x in xx]

plt.figure(figsize=(5,5))
plt.plot(xx, y)
plt.axis("equal")
plt.xlim(0, 1)
plt.ylim(0,1)
plt.plot(xx, xx, '--')
```

```python
xx = np.linspace(0, 1, 1000)
y = [iterate(tent, 4, x) for x in xx]

plt.figure(figsize=(5,5))
plt.plot(xx, y)
plt.axis("equal")
plt.xlim(0, 1)
plt.ylim(0,1)
plt.plot(xx, xx, '--')
```

The $n$th iterate of the tent map has $2^n$ "pieces" (or "laps") with slopes of modulus $2^n$, and hence exactly $2^n$ points that satisfy $T^n(x)=x$.

The $i$th "piece" of the $n$th iterate (with $i=0, \ldots, 2^n-1$) has equation
$$T^n_i(x) =
\begin{cases}
2^n x-i, & \text{if $i$ is even and $\frac{i}{2^n} \le x < \frac{i+1}{2^n}$} \\
i+1 - 2^n x, & \text{if $i$ is odd and $\frac{i}{2^n} \le x < \frac{i+1}{2^n}$} \\
\end{cases}
$$

Thus the solution of $T^n_i(x) = x$ satisfies
$$x^n_i = 
\begin{cases}
\frac{i}{2^n - 1}, & \text{if $i$ is even}; \\
\frac{i+1}{1 + 2^n}, & \text{if $i$ is odd}.
\end{cases}
$$

```python
def tent_fixed_point(i, n):
    if i % 2 == 0: # even
        return float(i) / (2**n - 1)
    
    return float(i+1) / (1 + 2**n)
```

```python
xx = np.linspace(0, 1, 1000)

n = 4

y = [iterate(tent, n, x) for x in xx]

plt.figure(figsize=(5,5))
plt.plot(xx, y)
plt.axis("equal")
plt.xlim(0, 1)
plt.ylim(0,1)
plt.plot(xx, xx, '--')

fixed_points = [tent_fixed_point(i, n) for i in range(0, 2**n)]

plt.plot(fixed_points, [iterate(tent, n, x) for x in fixed_points], 'o')
    
```

## Homeomorphism of the tent map with $f_4$

For $x \in [0,1]$, it turns out that we can apply the change of variables 
$$x = h(y) = \sin^2(\textstyle \frac{\pi y} {2}).$$
 If $$x_{n+1} = f_4(x_n) = 4 x_n (1-x_n)$$
then
$$y_{n+1} = T(y_n).$$

We say that there is a *conjugacy*, given by
We have
$$x_{n+1} = f_4(x_n) = f_4(h(y_n)),$$
and also
$$x_{n+1} = h(y_{n+1}) = h(T(y_n)).$$
Thus
$$f_4 \circ h = h \circ T,$$
or
$$f_4 = h \circ T \circ h^{-1}.$$

Thus, iterating gives
$$f_4^n = h \circ T^n \circ h^{-1},$$
so that a point $y$ such that 
$$T^n(y) = y$$
may be mapped to a point $x = h(y)$ such that
$$f_4^n(x) = f_4^n(h(y)) = h(T^n(y)) = h(y) = x.$$



```python
def h(y):
    return np.sin(np.pi * y / 2.)**2
```
